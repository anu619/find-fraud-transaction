```python
# Import functions from user_statistics_module module
import user_statistics_module

# Case selection
case_details="Please choose any one of the options listed below\n" \
            "[1] Maximum and minimum amount a user transacted\n" \
            "[2] Centroid Transaction amount of user transaction\n" \
            "[3] Distance from Centroid\n" \
            "[4] Calculate Standard deviation\n" \
            "[5] Calculate Varience\n" \
            "[6] Fraudulent check\n" \
            "[7] Check the distance between two transactions of user\n" \
            "[8] Check the distance of transcation of two users\n" \
            "[9] Exit\n" \

#Main function
def main():
    loop = True
    while loop:
        try:
            case=int(input(case_details))
            #Check the selection in the range
            if case not in range(1,10):
                print("Please choose the option between 1 to 9")
                
            #Case 1 Calculate Maximum and minimum value
            elif case==1:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                max_transaction=user_statistics_module.max_min_amount_transaction(user_id)
                min_transaction=user_statistics_module.max_min_amount_transaction(user_id)
                print("Maximum Transaction is {}".format(round(max_transaction[0],2)))
                print("Minimum Transaction is {}".format(round(min_transaction[1],2)))
            #Case 2 Calculate centroid value
            elif case==2:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                cent=user_statistics_module.calculate_centroid(user_id)
                print("Centroid value is: {}".format(cent))
            #Case 3 Calculate Distance of transaction from centriod
            elif case==3:
                try:
                    user_details ="Please Enter the User ID : "
                    user_id=int(input(user_details))
                    dist=user_statistics_module.calculate_distance(user_id)
                    print("Distance of transaction from centriod: {}".format(dist))
                except:
                    "print Invalid Transaction ID!!"
            #Case 4 Calculate Standard Deviation of Transaction
            elif case==4:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                std_deviation_data=user_statistics_module.calculate_standarddeviation(user_id)
                print("Standard Deviation of Transaction is {}".format(std_deviation_data))
            #Case 5 Calculate Variance of Transaction
            elif case==5:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                calulate_variance_transaction=user_statistics_module.calulate_variance(user_id)
                print("Variance of Transaction is {}".format(calulate_variance_transaction))
            #Case 6 Find fraud Transaction
            elif case==6:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                user_statistics_module.check_fraud_transaction(user_id)
            #Case 7 Find distance between two Transaction
            elif case==7:
                user_details ="Please Enter the User ID : "
                user_id=int(input(user_details))
                user_statistics_module.distancebetweentwotransactions(user_id)
            #Case 8 Find distance of two user Transaction
            elif case==8:
                user_details ="Please Enter the First User ID : "
                user_id=int(input(user_details))
                user_statistics_module.distanceusertransactions(user_id)
            elif case==9:
                break
            else:
                break
        except ValueError:
            print("Please enter the interger value!")
            
        except:
            print("Unwanted error, Please try again")
                    

main()
```
